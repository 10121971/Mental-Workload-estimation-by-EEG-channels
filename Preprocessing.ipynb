```python
'''Preprocessing of data'''

import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Step 1: Load the EEG dataset
raw = mne.io.read_raw_edf(r"C:\Users\Bhavya Prakash\OneDrive\Desktop\INMAS, DRDO\eeg-during-mental-arithmetic-tasks-1.0.0\eeg-during-mental-arithmetic-tasks-1.0.0\Subject16_1.edf", preload=True)
raw.plot(duration=1, n_channels=5, scalings='auto')


# Reading different valuable info from the dataset
print(f"Number of channels: {raw.info['nchan']}")
print(f"Sampling frequency: {raw.info['sfreq']} Hz")
print(f"Data length: {raw.n_times} samples")
print(f"Channel names: {raw.ch_names}")
print(raw.get_channel_types())
print(f"Data range: {raw.get_data().min()} - {raw.get_data().max()}")


# Apply a bandpass filter (for example, between 1 Hz and 50 Hz)
raw.filter(0.1, 45)


# Apply ICA for artifact removal
ica = mne.preprocessing.ICA(n_components=21, random_state=0)
ica.fit(raw)
ica.apply(raw)


# Step 3: Referencing
# Re-reference to average reference
raw.set_eeg_reference('average')


# Step 4: Resampling
raw.resample(256)  # Resample to a lower sampling rate (e.g., 250 Hz)


# Loading raw data into DataFrame of pandas
data = raw.get_data()
channel_names = raw.ch_names
df = pd.DataFrame(data=data.T, columns=channel_names)


# Saving preprocessed data to a csv file for each subject
csv_file = r'C:\Users\Bhavya Prakash\Downloads\EEG project'
df.to_csv(csv_file, index=False)
pd.read_csv(csv_file)


# Visualize the EEG data
raw.plot(duration=1, n_channels=5, scalings='auto')


# Reading different valuable info from the preprocessed dataset
print(f"Number of channels: {raw.info['nchan']}")
print(f"Sampling frequency: {raw.info['sfreq']} Hz")
print(f"Data length: {raw.n_times} samples")
print(f"Channel names: {raw.ch_names}")
print(raw.get_channel_types())


print(f"Filtering information: {raw.info['highpass']} Hz - {raw.info['lowpass']} Hz")
if raw.annotations:
    print("Annotations found:")
    print(raw.annotations.description)
else:
    print("No annotations found.")
    
    
print(f"Data range: {raw.get_data().min()} - {raw.get_data().max()}")
```
